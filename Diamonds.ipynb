Histograms are representing the count based on each of the characteristics. 

```python
def graph_histogram(column_name, color_name): 
  import seaborn as sns

  sns.set()

  import matplotlib.pyplot as plt

  plt.figure(figsize=(8,5))

  #x value
  plt.xlabel(column_name)
  #y value
  plt.ylabel("Count") 

  #description 
  plt.title(f'{column_name} Histogram', size=16)

  plt.hist(df[column_name],color=color_name,alpha=0.5)
  plt.show()

```

```python
#carat
graph_histogram('carat', 'navy')
```

```python
#cut
graph_histogram('cut', 'green')
```

```python
#color
graph_histogram('color', 'pink')
```

```python
#clarity
graph_histogram('clarity', 'yellow')
```

#Cross Validation 

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor, XGBRFRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
def cross_val(model, cv=10):
  neg_mean_squared_scores = cross_val_score(model, X, y, cv=cv, 
                           scoring='neg_mean_squared_error')
  rmse = (-neg_mean_squared_scores)**0.5
  return rmse.mean()
```

```python
cross_val(LinearRegression()) 
```

```python
cross_val(DecisionTreeRegressor())
```

```python
cross_val(RandomForestRegressor())
```

```python
cross_val(XGBRegressor())
```

```python
cross_val( XGBRFRegressor()) 
```

```python
cross_val(KNeighborsRegressor())
```

#Hyperparameter Fine-Tuning

```python
cross_val(XGBRegressor(objective='reg:squarederror', 
                       max_depth=5,
                       n_estimators=500,
                       subsample=0.5))
```

```python
cross_val(DecisionTreeRegressor(max_depth=12))
```

```python
cross_val(RandomForestRegressor(max_depth=6, n_estimators=200,
                                min_samples_split=6))
```

GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

xgb = XGBRegressor(objective='reg:squarederror')
params = {'max_depth':[2, 3, 5, 6, 8]}

# Instantiate GridSearchCV as grid_reg
grid = GridSearchCV(xgb, params)

# Fit grid_reg on X_train and y_train
grid.fit(X, y)

# Extract best params
best_params = grid.best_params_

# Print best params
print("Best params:", best_params)

# Compute best score
best_score = grid.best_score_

# Print best score
print("Best score: {:.5f}".format(best_score))
```
